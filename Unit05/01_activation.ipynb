## sigmoid 
$$ 1/(1+e^{-z}) $$

````python
import numpy as np
import matplotlib.pylab as plt
import time

def sigmoid(z):
    return 1 / ( 1  + np.exp(-z) )
    
````

````python
# -5~5中間每隔0.1，創建一個點
start = time.time()
X = np.arange(-5.0, 5.0, 0.1)
X
````

````python
Y = sigmoid(X)
Y
````

````python
start = time.time()
X = np.arange(-5.0, 5.0, 0.1)
Y = sigmoid(X)
now = time.time()
plt.plot(X, Y)
plt.ylim(-0.1, 1.1)
plt.show()
print("It cost %f sec" % (now-start))
````

## Tangent
$$ -1\leq  tanh(z)\leq 1$$

公式：
$$z=tanh(z)=\frac{e^{z}-e^{-z}}{e^{z}+e^{-z}} \quad$$
    
優點就是與 sigmoid 的區別是，tanh 解決了sigmoid的輸出非「零為中心」的問題，因此實際應用中 tanh 會比 sigmoid 更好。<br>

缺點類似sigmoid，雖然收斂速度相對快了，但也存在梯度彌散，而且也有冪計算，相對耗時。

````python
def tangent(x):
    return (np.exp(x)-np.exp(-x)) / (np.exp(x) + np.exp(-x))

````

````python
start = time.time()
x = np.arange(-5.0, 5.0, 0.1)
y = tangent(x)
now = time.time()
plt.plot(x, y)
plt.ylim(-1.1, 1.1)
plt.show()
print("It cost %f sec" % (now-start))
````

## Relu
$$relu=max(0,z)$$
優點：
* SGD算法的收斂速度比 sigmoid 和 tanh 快；（梯度不會飽和，解決了梯度消失問題）
* 計算復雜度低，不需要進行指數運算；
* 適合用於後向傳播。

缺點：

* ReLU的輸出不是zero-centered；
* Dead ReLU Problem（神經元壞死現象）：某些神經元可能永遠不會被激活，導致相應參數永遠不會被更新（在負數部分，梯度為0）。產生這種現象的兩個原因：參數初始化問題；learning rate太高導致在訓練過程中參數更新太大。 避免將learning rate設置太大或使用自動調節learning rate的算法。

````python
def relu(x):
    return np.maximum(0,x)
    '''
    for i in range(len(x)):
        if x[i] < 0:
            x[i]=0
    return x
    '''
````

````python
start = time.time()
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
now = time.time()
plt.plot(x, y)
plt.ylim(-1.0, 5.5)
plt.show()
print("It cost %f sec" % (now-start))
````
